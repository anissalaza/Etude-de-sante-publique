### Question n°3 - Calculez la disponibilité alimentaire en kcal puis en kg de protéines

```python
# On nettoie le df population pour pouvoir le joindre au df bilan_alim
popul = pop.drop(columns=["Élément","Produit","Code Domaine","Domaine","Code Élément",
                          "Code Produit","Code année","Année","Unité","Symbole",
                          "Description du Symbole","Zone"])
popul=popul.rename(columns={"Valeur":"Population"})
popul.head()
```

```python
# On joint les df
bilan_alim_pop = pd.merge(bilan_alim, popul, on="Code zone", how="left")
bilan_alim_pop.head()
```

```python
# On supprime les colonnes dont on n'a pas besoin
bilan_alim_popu = bilan_alim_pop.drop(columns=[
                    "Disponibilité de matière grasse en quantité (g/personne/jour)",
                                                           "Semences","Traitement",
                                                            "Variation de stock"])

# On renomme les colonnes pour que ce soit plus intuitif
bilan_alim_popu = bilan_alim_popu.rename(columns={
                    "Disponibilité alimentaire (Kcal/personne/jour)":"dispo_alim_kcal_j",
            "Disponibilité de protéines en quantité (g/personne/jour)":"dispo_prot_j",
    "Population":"population",
    "Disponibilité alimentaire en quantité (kg/personne/an)":"dispo_alim_kg",
    "Zone_x":"zone", "Disponibilité intérieure":"dispo_int", "Nourriture":"nourriture"})
bilan_alim_popu.head()
```

###### a/ Disponibilité alimentaire en kcal

```python
# On ajoute une colonne contenant la disponibilité alimentaire en kcal
bilan_alim_popu["bilan_alim_kcal"] = (bilan_alim_popu["dispo_alim_kcal_j"]*
                                      (bilan_alim_popu["population"]*1000)*365)
bilan_alim_popu.head()
```

###### b/ Disponibilité alimentaire en kg de protéines

```python
# On ajoute la colonne contenant la disponibilité alimentaire en kg
bilan_alim_popu["bilan_alim_kg_prot"] = ((bilan_alim_popu["dispo_prot_j"]/1000)*
                                         (bilan_alim_popu["population"]*1000)*365)
bilan_alim_popu.head()
```

### Question n°4 - Calculez pour chaque produit le ratio énergie/poids en kcal/kg

```python
# Définitions des fonctions pour les calculs des ratios

def calcul_ratio_kcal(row):
    if (row["bilan_alim_kcal"]!=0) & (row["nourriture"]!=0):
        ratio = row["bilan_alim_kcal"]/(row["nourriture"]*1e6)
    else: 
        ratio = math.nan
    return ratio

def calcul_ratio_prot(row):
    if (row["bilan_alim_kg_prot"]!=0) & (row["nourriture"]!=0):
        ratio = row["bilan_alim_kg_prot"]/(row["nourriture"]*1e6)*100
    else: 
        ratio = math.nan
    return ratio
```

```python
# Calcul du ratio énergie/poids en kcal/kg

# On ajoute la colonne au df
bilan_alim_popu["ratio_energie"]= bilan_alim_popu.apply (
    lambda row: calcul_ratio_kcal(row), axis=1)

bilan_alim_popu.head()
```

```python
# Vérification avec les oeufs (sur Wikipédia 147kcal/100g donc 1470kcal/kg)
ratio_oeufs = bilan_alim_popu.loc[
                                (bilan_alim_popu["Produit"]=="Oeufs") 
                                  & (bilan_alim_popu["ratio_energie"]!=0) 
                                  & (bilan_alim_popu["ratio_energie"]!=np.inf), 
                                ["ratio_energie"]].mean()
print("La valeur calorique d'un oeuf est ici de",round(ratio_oeufs, 2).values[0],"/kg soit 134,5/100g. A titre de comparaison, Wikipédia donne pour exemple une valeur calorique de 147kcal/100g.")
```

```python
# Calcul du pourcentage de protéines de chaque produit pour chaque pays

# On ajoute la colonne pourc_prot

bilan_alim_popu["pourc_prot"]= bilan_alim_popu.apply (
    lambda row: calcul_ratio_prot(row), axis=1)

bilan_alim_popu.head()
```

```python
# Vérification avec l'avoine (10,7g pour 100g sur Wikipédia)
avoine = bilan_alim_popu.loc[
                                (bilan_alim_popu["Produit"]=="Avoine") 
                                  & (bilan_alim_popu["pourc_prot"]!=0) 
                                  & (bilan_alim_popu["pourc_prot"]!=np.inf), 
                                ["pourc_prot"]].mean()
print("La teneur en protéines de l'avoine est ici de",round(avoine, 2).values[0],"g. A titre de comparaison, Wikipédia donne pour exemple une teneur en protéines de 10,7g/100g.")
```

```python
# Les 15 produits les plus exportés par les pays victimes de sous-alimentation
bm_exp = bm_exp_imp[["Produit", 'Exportations - Quantité']]\
.groupby(by=['Produit']).sum().reset_index()\
.sort_values(by='Exportations - Quantité', ascending=False)

produits_export_par_pays_sous_alim = bm_exp.Produit[:15]
produits_export_par_pays_sous_alim
```

*partie I - b/ Sélectionnez les 200 plus grandes importations de ces produits. Groupez ces importations par produit, afin d'avoir une table contenant 1 ligne pour chacun des 15 produits.*

```python
# Sélection des 200 plus grandes importations
bm_imp = bm_sous_nut.loc[
    bm_sous_nut['Produit'].isin(produits_export_par_pays_sous_alim),]\
.sort_values('Importations - Quantité', ascending=False)\
[:200]
bm_imp.head()
```

*partie I - c/ Ensuite, calculez pour chaque produit les 2 quantités suivantes :*
- le ratio entre la quantité destinée aux "Autres utilisations" (Other uses) et la disponibilité intérieure.
- le ratio entre la quantité destinée à la nourriture animale et la quantité destinée à la nourriture (animale + humaine)

```python
# On sélectionne les colonnes dont on a besoin
ratios = bm_imp[['Zone', 'Produit','Autres utilisations',
               'dispo_int','Aliments pour animaux','nourriture']]
# On effectue le groupby
ratios_sous_alim = ratios.groupby('Produit').sum().reset_index()
ratios_sous_alim.head()
```

```python
# On effectue les calculs pour chacun des ratios

# Ratio n°1 entre la quantité destinée aux Autres Utilisations et la disponibilité intérieure
ratios_sous_alim['ratio_AU_dispo_int'] = (ratios_sous_alim['Autres utilisations'] 
                                              / ratios_sous_alim['dispo_int'])

# Ratio n°2 entre la quantité destinée à la nourriture animale et la nourriture globale
ratios_sous_alim['ratio_al_anim_nourr'] = (ratios_sous_alim['Aliments pour animaux'] 
                                             / (ratios_sous_alim['Aliments pour animaux'] 
                                                   + ratios_sous_alim['nourriture']))

# On affiche le dataframe avec les produits et les ratios
ratios_sous_alim = ratios_sous_alim[['Produit', 'ratio_AU_dispo_int', 
                                                                 'ratio_al_anim_nourr']]
ratios_sous_alim.head()
```

*partie II - Donnez les 3 produits qui ont la plus grande valeur pour chacun des 2 ratios*

```python
# Les 3 produits qui ont la plus grande valeur pour le ratio 1
top3_ratio_AU_dispo_int = ratios_sous_alim[['Produit', 'ratio_AU_dispo_int']].\
sort_values('ratio_AU_dispo_int', ascending=False)[:3]
top3_ratio_AU_dispo_int
```

```python
print(">> Les produits importés par les pays victimes de sous-alimentation qui ont le ratio autres utilisations / disponibilité intérieure le plus élevé sont :",top3_ratio_AU_dispo_int['Produit'].values.tolist())
```

```python
# Les 3 produits qui ont la plus grande valeur pour le ratio 2
top3_ratio_al_anim_nourr = ratios_sous_alim[['Produit', 'ratio_al_anim_nourr']]\
.sort_values('ratio_al_anim_nourr', ascending=False)[:3]
top3_ratio_al_anim_nourr
```

```python
print(">> Les produits importés par les pays victimes de sous-alimentation qui ont le ratio nourriture animale / nourriture globale le plus élevé sont :",top3_ratio_al_anim_nourr['Produit'].values.tolist())
```
